Filtering data is a fundamental task in programming. This Python script will demonstrate how to filter a list of dictionaries (a common way to represent tabular data, similar to a JSON array or a CSV file) based on various conditions.

We'll cover:
1.  **Filtering by exact match.**
2.  **Filtering by numeric comparison (>, <, >=, <=).**
3.  **Filtering by string operations (contains, starts with).**
4.  **Filtering with multiple conditions (AND, OR logic).**

### Python Script: Data Filter

```python
import json

def filter_data(data, conditions, logical_operator="AND"):
    """
    Filters a list of dictionaries based on one or more conditions.

    Args:
        data (list of dict): The list of dictionaries to filter.
        conditions (list of dict): A list of dictionaries, where each dictionary
                                   defines a single filtering condition.
                                   Example: `{"field": "age", "operator": ">", "value": 30}`
                                   Supported operators: "==", "!=", ">", "<", ">=", "<=",
                                   "contains", "starts_with", "ends_with", "in" (for lists).
        logical_operator (str): How to combine multiple conditions. "AND" (default)
                                or "OR". Case-insensitive.

    Returns:
        list of dict: A new list containing only the dictionaries that match
                      the specified conditions.
    """
    if not conditions:
        return list(data) # No conditions, return all data

    filtered_results = []
    logical_operator = logical_operator.upper()

    for item in data:
        match_item = False

        if logical_operator == "AND":
            match_item = True # Assume true, prove false
            for condition in conditions:
                if not _evaluate_single_condition(item, condition):
                    match_item = False
                    break # One condition failed, so 'AND' fails for this item
        elif logical_operator == "OR":
            match_item = False # Assume false, prove true
            for condition in conditions:
                if _evaluate_single_condition(item, condition):
                    match_item = True
                    break # One condition passed, so 'OR' passes for this item
        else:
            print(f"Warning: Unsupported logical operator '{logical_operator}'. Defaulting to AND.")
            match_item = True # Assume true for AND fallback
            for condition in conditions:
                if not _evaluate_single_condition(item, condition):
                    match_item = False
                    break

        if match_item:
            filtered_results.append(item)

    return filtered_results

def _evaluate_single_condition(item, condition):
    """
    Helper function to evaluate a single condition for a given item.
    """
    field = condition.get("field")
    operator = condition.get("operator")
    value = condition.get("value")

    if field is None or operator is None or value is None:
        print(f"Warning: Invalid condition format: {condition}. Skipping.")
        return False

    item_value = item.get(field)

    # Handle cases where the item might not have the field
    if item_value is None:
        return False

    try:
        if operator == "==":
            return item_value == value
        elif operator == "!=":
            return item_value != value
        elif operator == ">":
            return item_value > value
        elif operator == "<":
            return item_value < value
        elif operator == ">=":
            return item_value >= value
        elif operator == "<=":
            return item_value <= value
        elif operator == "contains":
            return isinstance(item_value, str) and isinstance(value, str) and value.lower() in item_value.lower()
        elif operator == "starts_with":
            return isinstance(item_value, str) and isinstance(value, str) and item_value.lower().startswith(value.lower())
        elif operator == "ends_with":
            return isinstance(item_value, str) and isinstance(value, str) and item_value.lower().endswith(value.lower())
        elif operator == "in": # Checks if item_value is in a list (value should be a list)
            return item_value in value if isinstance(value, list) else False
        else:
            print(f"Warning: Unsupported operator '{operator}'. Skipping condition.")
            return False
    except TypeError as e:
        print(f"Error comparing values for field '{field}' with operator '{operator}': {e}. Skipping condition.")
        return False
    except Exception as e:
        print(f"An unexpected error occurred during condition evaluation: {e}. Skipping condition.")
        return False

# --- Sample Data ---
employees_data = [
    {"id": 1, "name": "Alice Smith", "department": "HR", "salary": 60000, "status": "active", "projects": ["P1", "P2"]},
    {"id": 2, "name": "Bob Johnson", "department": "IT", "salary": 75000, "status": "active", "projects": ["P3"]},
    {"id": 3, "name": "Charlie Brown", "department": "HR", "salary": 62000, "status": "inactive", "projects": ["P1"]},
    {"id": 4, "name": "David Lee", "department": "Sales", "salary": 80000, "status": "active", "projects": ["P4", "P5"]},
    {"id": 5, "name": "Eve Davis", "department": "IT", "salary": 70000, "status": "active", "projects": ["P3", "P6"]},
    {"id": 6, "name": "Frank White", "department": "Marketing", "salary": 65000, "status": "active", "projects": ["P7"]},
    {"id": 7, "name": "Grace Green", "department": "Sales", "salary": 90000, "status": "active", "projects": ["P4"]},
    {"id": 8, "name": "Heidi Grey", "department": "IT", "salary": 72000, "status": "inactive", "projects": ["P3"]},
    {"id": 9, "name": "Ivan Black", "department": "HR", "salary": 58000, "status": "active", "projects": ["P8"]},
]

print("--- Original Data ---")
print(json.dumps(employees_data, indent=2))
print("\n" + "="*50 + "\n")


# --- Filtering Examples ---

# 1. Filter by exact match (department = IT)
print("1. Employees in the 'IT' department:")
conditions_it = [{"field": "department", "operator": "==", "value": "IT"}]
filtered_it = filter_data(employees_data, conditions_it)
print(json.dumps(filtered_it, indent=2))
print("\n" + "-"*50 + "\n")

# 2. Filter by numeric comparison (salary > 70000)
print("2. Employees with salary greater than 70000:")
conditions_high_salary = [{"field": "salary", "operator": ">", "value": 70000}]
filtered_high_salary = filter_data(employees_data, conditions_high_salary)
print(json.dumps(filtered_high_salary, indent=2))
print("\n" + "-"*50 + "\n")

# 3. Filter by string operation (name contains 'smith', case-insensitive)
print("3. Employees whose name contains 'smith':")
conditions_name_contains = [{"field": "name", "operator": "contains", "value": "smith"}]
filtered_name_contains = filter_data(employees_data, conditions_name_contains)
print(json.dumps(filtered_name_contains, indent=2))
print("\n" + "-"*50 + "\n")

# 4. Filter with multiple conditions (department = HR AND status = active)
print("4. Active employees in 'HR' department (AND logic):")
conditions_hr_active = [
    {"field": "department", "operator": "==", "value": "HR"},
    {"field": "status", "operator": "==", "value": "active"}
]
filtered_hr_active = filter_data(employees_data, conditions_hr_active, logical_operator="AND")
print(json.dumps(filtered_hr_active, indent=2))
print("\n" + "-"*50 + "\n")

# 5. Filter with multiple conditions (department = Sales OR salary < 60000)
print("5. Employees in 'Sales' OR with salary less than 60000 (OR logic):")
conditions_sales_or_low_salary = [
    {"field": "department", "operator": "==", "value": "Sales"},
    {"field": "salary", "operator": "<", "value": 60000}
]
filtered_sales_or_low_salary = filter_data(employees_data, conditions_sales_or_low_salary, logical_operator="OR")
print(json.dumps(filtered_sales_or_low_salary, indent=2))
print("\n" + "-"*50 + "\n")

# 6. Filter for items where a list field contains a specific value (e.g., project 'P1')
print("6. Employees involved in 'P1' project:")
conditions_project_p1 = [{"field": "projects", "operator": "contains", "value": "P1"}]
filtered_project_p1 = filter_data(employees_data, conditions_project_p1)
print(json.dumps(filtered_project_p1, indent=2))
print("\n" + "-"*50 + "\n")

# 7. Filter for items where a value is in a list of allowed values
print("7. Employees in 'HR' or 'Marketing' departments:")
conditions_dept_in_list = [{"field": "department", "operator": "in", "value": ["HR", "Marketing"]}]
filtered_dept_in_list = filter_data(employees_data, conditions_dept_in_list)
print(json.dumps(filtered_dept_in_list, indent=2))
print("\n" + "="*50 + "\n")
```

```python
import pandas as pd
from typing import List, Union

def filter_employees_by_age_and_department(df: pd.DataFrame) -> pd.DataFrame:
    """
    Filters an employee DataFrame to include only employees older than 30 in the Engineering department.

    This function expects the input DataFrame to contain at least 'Age' and 'Department' columns.

    Args:
        df: The input pandas DataFrame containing employee data.
            Expected columns:
            - 'Age' (numeric type): Represents the employee's age.
            - 'Department' (string/object type): Represents the employee's department.

    Returns:
        A new pandas DataFrame containing only the rows that satisfy both filtering conditions:
        'Age' > 30 AND 'Department' == 'Engineering'.
        Returns an empty DataFrame with the original columns if the input is empty or
        if no rows match the criteria.

    Raises:
        ValueError: If 'Age' or 'Department' columns are missing from the input DataFrame.
    """
    # 1. Handle empty input DataFrame gracefully
    if df.empty:
        # Return an empty DataFrame with the same columns as the input,
        # which is consistent with how pandas filtering would behave.
        return df.copy()

    # 2. Validate required columns for robustness
    required_columns: List[str] = ['Age', 'Department']
    missing_columns: List[str] = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        raise ValueError(
            f"Input DataFrame is missing required columns: {', '.join(missing_columns)}. "
            "Please ensure 'Age' and 'Department' columns are present."
        )

    # 3. Apply filtering conditions using boolean indexing
    # Ensure 'Age' column is treated numerically for comparison.
    # If 'Age' contains non-numeric values, this will cause issues.
    # For production-ready code, consider explicit type conversion with error handling
    # if data quality is not guaranteed (e.g., df['Age'] = pd.to_numeric(df['Age'], errors='coerce'))
    # For this function, we assume 'Age' is already numeric or compatible.
    
    age_greater_than_30 = df['Age'] > 30
    is_engineering_department = df['Department'] == 'Engineering'

    # Combine conditions using the bitwise AND operator (&)
    filtered_employees_df = df[age_greater_than_30 & is_engineering_department]

    return filtered_employees_df

if __name__ == "__main__":
    # --- Example Usage ---

    # 1. Create a sample DataFrame for demonstration
    employee_data = {
        'EmployeeID': [1, 2, 3, 4, 5, 6, 7, 8, 9],
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan'],
        'Age': [25, 32, 40, 29, 35, 30, 45, 28, 50],
        'Department': ['HR', 'Engineering', 'Engineering', 'Sales', 'Engineering', 'HR', 'Marketing', 'Engineering', 'Engineering'],
        'Salary': [60000, 80000, 95000, 70000, 90000, 62000, 100000, 75000, 110000]
    }
    employees_df = pd.DataFrame(employee_data)

    print("--- Original Employee DataFrame ---")
    print(employees_df)
    print("\n" + "="*50 + "\n")

    # 2. Test Case: Valid Filtering
    print("--- Filtered Employees (Age > 30, Department == 'Engineering') ---")
    try:
        filtered_results = filter_employees_by_age_and_department(employees_df)
        print(filtered_results)
    except ValueError as e:
        print(f"Error during filtering: {e}")
    print("\n" + "="*50 + "\n")

    # 3. Test Case: Empty Input DataFrame
    print("--- Test with Empty DataFrame ---")
    empty_employees_df = pd.DataFrame(columns=['EmployeeID', 'Name', 'Age', 'Department', 'Salary'])
    try:
        filtered_empty_results = filter_employees_by_age_and_department(empty_employees_df)
        print("Filtered Empty DataFrame:\n", filtered_empty_results)
    except ValueError as e:
        print(f"Error during filtering: {e}")
    print("\n" + "="*50 + "\n")

    # 4. Test Case: DataFrame with no matching rows
    print("--- Test with DataFrame containing no matching rows ---")
    no_match_data = {
        'EmployeeID': [10, 11],
        'Name': ['Liam', 'Mia'],
        'Age': [28, 25],
        'Department': ['Engineering', 'Sales']
    }
    no_match_df = pd.DataFrame(no_match_data)
    try:
        filtered_no_match_results = filter_employees_by_age_and_department(no_match_df)
        print("Filtered (no matches):\n", filtered_no_match_results)
    except ValueError as e:
        print(f"Error during filtering: {e}")
    print("\n" + "="*50 + "\n")

    # 5. Test Case: Missing 'Age' column
    print("--- Test with DataFrame missing 'Age' column ---")
    df_missing_age = employees_df.drop(columns=['Age'])
    try:
        filter_employees_by_age_and_department(df_missing_age)
    except ValueError as e:
        print(f"Caught expected error: {e}")
    print("\n" + "="*50 + "\n")

    # 6. Test Case: Missing 'Department' column
    print("--- Test with DataFrame missing 'Department' column ---")
    df_missing_department = employees_df.drop(columns=['Department'])
    try:
        filter_employees_by_age_and_department(df_missing_department)
    except ValueError as e:
        print(f"Caught expected error: {e}")
    print("\n" + "="*50 + "\n")
```
